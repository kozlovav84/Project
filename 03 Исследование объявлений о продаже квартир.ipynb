## Предобработка данных

### Определяем и изучаем пропущенные значения

Для определения и изучения пропущенных значений воспользуемся методами `unique()` и `value_counts()`. Чтобы не повторять код много много раз, поместим его в цикл. 

```python
for column in data.columns:
    print('Данные по столбцу:', column)
    print(data[column].unique(), '\n')
    print(data[column].value_counts(), '\n')
    print ('-' * 100) #отделем визуально данные одного вывода от другого
```

Вызывают интерес следующие столбцы:
- `data['rooms']`: на первый взгляд, количество комнат "0" выглядит странным. Но если предположить, что так обозначали студии, то всё становится на свои места. Хорошо бы уточнить этот факт у того, кто готовил данные для анализа.В нашем исследовании принимаем, что "0" обозначает студии и не изменяем это значение;
- `data['ceiling_height']` встречается "nan", берём столбец в обработку;
- `data['floors_total']` встречается "nan", берём столбец в обработку. Хотя как обрабатывать пропущенные значения количества этажей не очень понимаю, тут будет уместным, СПРОСИТЬ У ПРЕПОДА !!!!!
- в `data['balcony']` встречается "nan". Принимаем в данном случае, что если значение не значение не заполнено, то скорее всего балконов нет. Берём столбец в обработку.

Начинаем со столбца "высота потолков", считаем количество пропущенных значений методом `insa()`:

```python
data[data['ceiling_height'].isna()]
```

Как я уже писал выше, минимальную высоту потолков можно принять 2,4 метра. С максимальной сложнее, например  <a href="https://spb.cian.ru/sale/flat/218657555/">в известной нам квартире</a> высота потолка достигает 5,1 м. Принимаем это значение за возможный максимум. Данные по высотам потолка нам потребуются для:
- отсеивания "выбросов";
- заполнения пропущенных значений.

Поскольку высота 5,1 метр больше исключение, чем правило, то чтобы это значение не повлияло на остальные, заполнять пропуски будем медианой.

```python
ceiling_height_median = (
    data.loc[(data.loc[:, 'ceiling_height'] <= 5.1) & 
             (data.loc[:, 'ceiling_height'] >= 2.4)]['ceiling_height'].median() 
                        ) #выбираем промежуток для медианы, используем пренос для простоты проверки и удобства 
ceiling_height_median #медианное значение высоты потолков, по факту не сильно отличается от среднего.
```

```python
data['ceiling_height'] = data['ceiling_height'].fillna(ceiling_height_median) #заполняем пропуски в колонке data['ceiling_height']
data = data.query('2.4 <=ceiling_height <= 5.1') #осекаю аномальные значения менее 2,4 м и более 5,1 м
data.info()
print ('-' * 100)
data['ceiling_height'].describe()
```

Продолжаем со столбцом "Балконы"

```python
data['balcony'] = data['balcony'].fillna(0) #пропуски в колонке с количеством балконов заполняю 0(как предлагалось сделать в описании проекта)
data.info()
```

Продолжаем со столбцом "Всего этажей в доме". 

```python
data[data['floors_total'].isna()] #Проверяю количество значений "NaN"
```

Всего 86 строк. Я долго ломал голову, как именно заполнить "NaN" в этом столбце. Посоветовавшись с преподавателем, решили заполнить служебным значением. Служебное значение принимаю 60, максимальное значение в данном столбце. Оно не соответствует реальноости и в последствии будет исключено, а так как таких строк немного, менее 1%,  то на результат исследования эта замена не повлияет.

```python
data['floors_total'] = data['floors_total'].fillna(60) #пропуски в колонке с количеством этажей заполняем 60
data.info()
```

Далее в столбце `data['floors_total']` отсекаем значения менее 35:

```python
data = data.query('floors_total <= 35') #напомним, что самый высокий дом СПб 35 этажей
data.describe()
```

На этом с пропусками всё. Пропуски скорее всего появились по техническим причинам при форматировании, выгрузке данных. Так же вероятно, что данные не были заполнены изначально. 

В таблице встречаются столбцы, которые очень логично привести к типу int64. К ним относятся: 
- `data['floors_total']`;
- `data['balcony']`.

Тип нужно поменять, так как не бывает 1,3 балкона и 5,86 этажей. Эти данные могут быть только целыми числами. Дополнительно приведём к тиу int64 данные столбца `data['last_price']`, так как нам не очень нужны копейки, да и в значении после запятой всегда ноль (смело избавимся от него).
Тип данных будем менять, применяя метод astype()

```python
data[['floors_total', 'balcony', 'last_price']] = data[['floors_total', 'balcony', 'last_price']].astype('int64')
data.info()
```

Петерь поменяем тип "object" на тип "datatime64" в столбце `data['first_day_exposition']`, применяя метод `to_datetime()`. Делаем это для того, чтобы в последствии добавить в таблицу новые столбцы для работы с датами.  

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'],format='%Y.%m.%dT%H:%M:%S')
data.info()
```

## Расчёты и добавление результатов в таблицу

### Считаем и добавляем в таблицу цену квадратного метра

Для того, чтобы определить цену квадратного метра нужно стоимость помещения разделить на площадь помещения. Новый стобец назовём `data['price_square_meter']`.

```python
data['price_square_meter'] = data['last_price'] / data['total_area']
data.head()
```

### Считаем и добавляем в таблицу день недели, месяц и год публикации объявления

```python
data['weekday'] = data['first_day_exposition'].dt.weekday # создаём дни недели: 0 - понедельник, 6 - воскресенье
data['month'] = data['first_day_exposition'].dt.month # создаём месяцы: 1 - январь, 12 - декабрь
data['year'] = data['first_day_exposition'].dt.year # создаём год.
data.head()
```

### Считаем и добавляем в таблицу этаж квартиры; варианты — первый, последний, другой

Для выполнения этой задачи будем использоватьметод категоризации.  Реализовать его нам поможет функция, а для того, чтобы получить столбец с группой нам необходимо воспользоваться методом `apply()`

```python
def type_floor(floor): #создаем функцию для определения категории этажности
    if floor['floor'] == 1:
        return 'первый'
    elif floor['floor'] == floor['floors_total']:
        return 'последний'
    else:
        return 'другой'
data['category_floor'] = data.apply(type_floor, axis=1) #результат работы функции добавляем в новый столбец 
data.head()
```

### Считаем и добавляем в таблицу соотношение жилой и общей площади, а также отношение площади кухни к общей

Для того, чтобы посчитать соотношение жилой и общей площади нужно жилую площадь разделить на общую площадь. 

```python
data['ratio_living_total'] = data['living_area'] / data['total_area'] # добавляем столдбец в таблицу
data.head()
```

Для того, чтобы посчитать отношение площади кухни к общей площади нужно площадь кухни разделить на общую площадь. Сделаем это скорее:

```python
data['ratio_kitchen_total'] = data['kitchen_area'] / data['total_area']
data.head()
```

Выделим квартиры в Санкт-Петербурге (`data['locality_name']`). Наша задача — выяснить, какая область входит в центр. Создадим столбец с расстоянием до центра в километрах: округлим до целых значений. После этого посчитаем среднюю цену для каждого километра. Построим график, который будет показывать, как цена зависит от удалённости от центра.

Выведем список характерных значений столбца расстояние до центра города `data['cityCenters_nearest']`:

```python
data['cityCenters_nearest'].describe()
```

Расстояние указано в метрах, но нам нужно определить его в километрах и округлить до целых значений. Создаём новый столбец:

```python
data['cityCenters_nearest_km'] = data['cityCenters_nearest'] / 1000 #создём новый столбец с расстоянием до центра в км
data.loc[data['cityCenters_nearest_km'].notnull(), 'cityCenters_nearest_km'] = (
    data.loc[data['cityCenters_nearest_km'].notnull(), 'cityCenters_nearest_km'].astype('int64')
                                                                            ) #округлем до целых значений
data['cityCenters_nearest_km'].describe()
```

Отлично, расстояние в метрах и округлено до целых значений (среднее и среднее отклонение не в счёт). 

Важное замечание: В данном расчёте я не буду брать значения Шушары, Парголово, Колпино, Пушкин и другие названия, входящие в состав Санкт-Петербурга, так как расстояние до центра СПб от недвижимости в них расположенных явно превышает среднее по столбцу `data['cityCenters_nearest_km']`.

Далее, выделяем квартиры в Санкт-Петербурге, выясняем, какая область входит в центр, и строим график.

```python
data_spb = data.query('locality_name == "Санкт-Петербург"')
data_km_mean = (
    data_spb.pivot_table(index='cityCenters_nearest_km', values='last_price', aggfunc='mean')
            ) #считаем среднюю цену для каждого км
data_km_mean.plot(kind='bar', y='last_price', grid=True, figsize=(12, 6)) #строим график
```

На графике чётко видно, что между 7 и 9 километров цена различается более чем на 10**6 рублей. Одна из наших задач - выяснить, какая область входит в центр. В центр входят квартиры в радиусе 8 км включительно.
